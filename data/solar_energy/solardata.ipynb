```python
# -----------------------------------------
# YData Profiling for Hourly Solar Dataset
# -----------------------------------------

import pandas as pd
from ydata_profiling import ProfileReport

# -----------------------------------------
# Load dataset
# -----------------------------------------
FILE_PATH = "HOURLY_SOLAR_DATA_IST(2).csv"
df = pd.read_csv(FILE_PATH)

print("✅ Hourly solar dataset loaded successfully!")
print("Shape:", df.shape)

# -----------------------------------------
# Optional: basic cleaning before profiling
# -----------------------------------------
# Remove physically invalid values (if any)
df = df[df["solar_kwh"] >= 0]
df = df[df["ALLSKY_SFC_SW_DWN_HOURLY"] >= 0]

# -----------------------------------------
# FULL PROFILING REPORT
# -----------------------------------------
print("\n🔍 Generating full profiling report...")

profile_full = ProfileReport(
    df,
    title="Hourly Solar Generation (IST) – Full Profile",
    explorative=True,
    minimal=True   # keeps report lighter & faster
)

profile_full.to_file("hourly_solar_full_profile.html")

print("✅ Full profiling saved as hourly_solar_full_profile.html")

# -----------------------------------------
# TARGET-FOCUSED PROFILING (solar_kwh)
# -----------------------------------------
print("\n🎯 Generating target-only profiling report...")

target_df = df[["solar_kwh"]]

profile_target = ProfileReport(
    target_df,
    title="Target Variable Profiling – solar_kwh",
    explorative=True,
    minimal=True
)

profile_target.to_file("solar_kwh_profile.html")

print("✅ Target profiling saved as solar_kwh_profile.html")

# -----------------------------------------
# LIGHTWEIGHT NUMERICAL SUMMARY (CSV)
# -----------------------------------------
print("\n📊 Generating lightweight statistical summary...")

summary = df.describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).transpose()
summary.to_csv("hourly_solar_summary.csv")

print("✅ Summary saved as hourly_solar_summary.csv")

print("\n🎉 YData profiling completed successfully!")
```

```python
# -----------------------------------------
# Pure Preprocessing for Hourly Solar Data
# -----------------------------------------

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("HOURLY_SOLAR_DATA_IST(2).csv")
print("✅ Dataset loaded:", df.shape)

# Drop column not needed for modeling
df = df.drop(columns=["YEAR"])

df["is_day"] = (df["ALLSKY_SFC_SW_DWN_HOURLY"] > 0).astype(int)

# Cyclical encoding for time features
df["hour_sin"] = np.sin(2 * np.pi * df["HOUR"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["HOUR"] / 24)

df["month_sin"] = np.sin(2 * np.pi * df["MONTH"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["MONTH"] / 12)

# Drop raw cyclical columns
df = df.drop(columns=["HOUR", "MONTH"])

# Final sanity check
print("✅ Final columns:")
print(df.columns.tolist())

# Save preprocessed dataset
OUTPUT_FILE = "HOURLY_SOLAR_DATA_PREPROCESSED(2).csv"
df.to_csv(OUTPUT_FILE, index=False)

print(f"\n🎉 Preprocessing done! Saved as {OUTPUT_FILE}")
```
