```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("haryana_hourly_generation_cleaned.csv")

# -----------------------------
# Handle datetime for ML
# -----------------------------
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract useful features
df['Year'] = df['datetime'].dt.year
df['Month'] = df['datetime'].dt.month
df['Day'] = df['datetime'].dt.day
df['Hour_of_day'] = df['datetime'].dt.hour
df['Weekday'] = df['datetime'].dt.weekday  # 0=Monday, 6=Sunday

# Drop original datetime column
df = df.drop(columns=['datetime'])

# -----------------------------
# Drop Coal (already in Coal_Gas)
# -----------------------------
df = df.drop(columns=['Coal'])

# -----------------------------
# Separate features and target
# -----------------------------
X = df.drop(columns=['Demand_MWh'])  # all features
y = df['Demand_MWh']                 # target variable

# -----------------------------
# Check the processed data
# -----------------------------
print("Features (X) columns:\n", X.columns)
print("\nTarget (y) preview:\n", y.head())
```

```python
# Combine features and target into one DataFrame
df_processed = X.copy()
df_processed['Demand_MWh'] = y  # add target back

# Save to new CSV
df_processed.to_csv("haryana_hourly_generation_preprocessed.csv", index=False)
print("✅ Preprocessed dataset saved as haryana_hourly_generation_preprocessed.csv")
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# Load raw file
# -----------------------------
df = pd.read_csv("haryana_hourly_generation_preprocessed.csv")

# Sort by time (VERY IMPORTANT)
df = df.sort_values(
    by=["Year", "Month", "Day", "Hour_of_day"]
).reset_index(drop=True)

# -----------------------------
# DROP non-meter features
# -----------------------------
drop_cols = [
    "Hydro",
    "Solar",
    "Renewables",
    "Solar Availability Flag",
    "Coal_Gas",
    "Ramp Rate (change in demand)",
    "Year"
]

df = df.drop(columns=drop_cols)

# -----------------------------
# Cyclical time encoding
# -----------------------------
df["hour_sin"] = np.sin(2 * np.pi * df["Hour_of_day"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["Hour_of_day"] / 24)

df["dow_sin"] = np.sin(2 * np.pi * df["Weekday"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["Weekday"] / 7)

# -----------------------------
# Lag features (PAST ONLY)
# -----------------------------
lags = [1, 2, 24, 168]
for lag in lags:
    df[f"Demand_lag_{lag}"] = df["Demand_MWh"].shift(lag)

# -----------------------------
# Rolling statistics (shifted)
# -----------------------------
df["Demand_roll_mean_6"] = df["Demand_MWh"].shift(1).rolling(6).mean()
df["Demand_roll_mean_24"] = df["Demand_MWh"].shift(1).rolling(24).mean()
df["Demand_roll_std_24"] = df["Demand_MWh"].shift(1).rolling(24).std()

# -----------------------------
# Trend features
# -----------------------------
df["Demand_diff_past_1h"] = df["Demand_lag_1"] - df["Demand_lag_2"]
df["Demand_diff_past_24h"] = df["Demand_lag_1"] - df["Demand_lag_24"]

# -----------------------------
# Drop rows with NaNs (from lagging)
# -----------------------------
df = df.dropna().reset_index(drop=True)

# -----------------------------
# Save clean dataset
# -----------------------------
df.to_csv("load_meter_ready_dataset.csv", index=False)

print("✅ Load-meter-ready dataset created")
```
